```python
Day = '15-1'

from common import read_daily_input
from itertools import zip_longest
import numpy as np
import pandas as pd

aoc_input = read_daily_input(Day)
```

```python
input_list = []

for j in aoc_input:
    sensor_x_val = int(j.split(':')[0].split(',')[0].split('=')[1])
    sensor_y_val = int(j.split(':')[0].split(',')[1].split('=')[1])
    beacon_x_val = int(j.split(':')[1].split(',')[0].split('=')[1])
    beacon_y_val = int(j.split(':')[1].split(',')[1].split('=')[1])
    
    input_list.append([[sensor_x_val,sensor_y_val], abs(sensor_x_val-beacon_x_val)+abs(sensor_y_val-beacon_y_val)])

df = pd.DataFrame(columns=['X','Y'])
```

```python
for sensor in input_list[0:1]:
    center = sensor[0]
    dist = sensor[1]
    
    for i in range(1,dist+1):
        x = list(range(center[0]-i,center[0]+i+1))
        x = x+list(reversed(x[1:-1]))
        y_vals = list(range(center[1]-i,center[1]+i+1))
        y = y_vals+list(reversed(y_vals[1:-1]))
        for n in range(i):
            val = y.pop(0)
            y.append(val)
        for k in range(len(x)):
            df = df.append({'X': x[k], 'Y': y[k]}, ignore_index=True)
```
